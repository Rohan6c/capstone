```python
with open('newyork_data.json') as json_data:
    newyork_data = json.load(json_data)
```

```python
newyork_data
```

```python
neighborhoods_data = newyork_data['features']
```

```python
neighborhoods_data[0]
```

```python
columns=['Borough','Neighbrhood','Latitude','Longitude']
neighborhoods=pd.DataFrame(columns=columns)
neighborhoods.head()
```

```python
for data in neighborhoods_data:
    Borough=data['properties']['borough']
    Neighbrhood=data['properties']['name']
    Location=data['geometry']['coordinates']
    Latitude=Location[1]
    Longitude=Location[0]
    neighborhoods=neighborhoods.append({'Borough':Borough,'Neighbrhood':Neighbrhood,'Latitide':Latitude,'Longitude':Longitude},ignore_index=True)
```

```python
neighborhoods=neighborhoods.drop('Latitude',1)
```

```python
neighborhoods.head()
```

```python
df=neighborhoods[neighborhoods['Borough']=='Manhattan']
df=df.reset_index()
df.head()
df=df.drop('index',1)
df.head()
```
