```python
def getnearbyvenues(names,latitude,longitude,radius=500):
    venuelist=[]
    for name,lat,lng in zip(names,latitude,longitude):
        print(name)
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lon, 
            radius, 
            limit)
        results=requests.get(url).json()["response"]['groups'][0]['items']
        venuelist.append([(name,lat,lon,v['venue']['name'],v['venue']['location']['lat'],v['venue']['location']['lng'], v['venue']['categories'][0]['name'])for v in results])
    venuenearby=pd.DataFrame(item for venueslist in venuelist for item in venueslist)
    venuenearby.columns=['Neighborhood Name','Neighborhood Latitude','Neighborhood Longitude','Venue Name','Venue Latitude','Venue Location','Venue Category']
    return(venuenearby)
```

```python
downtowntorontoonehot=pd.get_dummies(downtowntoronto['Venue Category'],prefix="",prefix_sep="")
downtowntorontoonehot['Neighrhood']=downtowntoronto['Neighborhood Name']
fixedcols=[downtowntorontoonehot.columns[-1]]+list(downtowntorontoonehot.columns[:-1])
torontocod=downtowntorontoonehot[fixedcols]
torontocod.head()
```

```python
torontogrp=torontocod.groupby('Neighrhood').mean().reset_index()
```

```python
torontogrp.head()
```

```python
five=5
for hood in torontogrp['Neighrhood']:
    print(hood)
    temp=torontogrp[torontogrp['Neighrhood']==hood].T.reset_index()
    temp=temp.iloc[1:]
    temp.columns=['Venue','Frequency']
    temp['Frequency']=temp['Frequency'].astype(float)
    temp=temp.round({'Frequency':2})
    print(temp.sort_values('Frequency',ascending=False).reset_index().head(five))
```
